### **Python Analysis and Visualization**

# **Sales and Pricing Analysis**

```python
# 1. Which brand sold the most cars in 2024?
brand_sales = carsdf["Brand"].value_counts().sort_values(ascending=True)

brand_sales.plot(
    kind="barh",
    title="Car Sales by Brand",
    color="lightblue",
    xlabel="Count",
    ylabel="Brand",
    figsize=(20,10)
)
plt.show()
print(brand_sales)
```

```python
# 2. What is the average final price of cars across different brands?
brand_avg_price = carsdf.groupby("Brand")["Final_Price"].mean().sort_values(ascending=False)
brand_avg_price = brand_avg_price.round(0).astype(int)

plt.figure(figsize=(20, 8))
sns.barplot ( x= brand_avg_price.index, y = brand_avg_price.values, color="green")
plt.title("Average Price by Brand")
plt.xlabel("Brand")
plt.ylabel("Average Price")
plt.show()

#brand_avg_price = brand_avg_price / 100000
print(brand_avg_price.round(2))
```

```python
# 3. Top 10 city generated the highest sales revenue?
city_revenue = carsdf.groupby("City")["Final_Price"].sum().sort_values(ascending=False).head(10)
city_revenue = city_revenue.round(0).astype(int)


plt.figure(figsize=(20, 10))
sns.lineplot (x= city_revenue.index, y = city_revenue.values,sizes=(20,20))
plt.title("Top 10 City by Revenue")
plt.xlabel("City")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()

print(city_revenue)
```

```python
# 4. Top 10 State generated the highest sales revenue?
state_revenue = carsdf.groupby("State")["Final_Price"].sum().sort_values(ascending=False).head(10)
state_revenue = state_revenue.round(0).astype(int)


plt.figure(figsize=(20, 10))
sns.barplot (x= state_revenue.index, y = state_revenue.values,color="darkblue")
plt.title("Top 10 State by Revenue")
plt.xlabel("State")
plt.ylabel("Revenue")
plt.show()

print(state_revenue)
```

```python
# 5.	How do ex-showroom prices compare to final prices across brands?
brand_price = carsdf.groupby("Brand")[["Ex_Showroom_Price",  "Final_Price"]].mean().sort_values( by= "Final_Price",ascending=False)
brand_price = brand_price.round(0).astype(int)

plt.figure(figsize = (20,8))
sns.barplot(data=brand_price,errorbar=None)
plt.title("Average Price by Brand")
plt.xlabel("Brand")
plt.ylabel("Average Price")
plt.show()

print(brand_price)
```

```python
# 6.Which car model had the highest discount offered?
max_discount_model = carsdf.groupby("Model")["Discount"].sum().sort_values(ascending=False).head()

plt.figure(figsize =(20,7))
plt.pie(max_discount_model)
plt.title("Max Discount Model")
plt.legend(max_discount_model.index, loc="best")
plt.show()

print(max_discount_model)
```

# **Customer Financing & Loan Analysis**

```python
# 7.	What percentage of cars were bought with a loan vs. full cash?
```

```python
# 8. Which brand has the highest average loan amount?
avg_loan_brand = carsdf.groupby("Brand")["Loan_Amount"].mean().sort_values(ascending=False).round(2).head()
print(avg_loan_brand)



plt.figure(figsize=(20, 7))
sns.scatterplot(data=avg_loan_brand)
plt.title("Average Loan Amount by Brand")
plt.xlabel("Brand")
plt.ylabel("Average Loan Amount")
plt.grid(True)
plt.show()
```

```python
# 9.What is the distribution of EMI months across buyers?
emi_months_dist = carsdf["EMI_Months"].value_counts().sort_index()
print(emi_months_dist)

plt.figure(figsize=(20,7))
plt.pie(emi_months_dist,autopct="%.2f%%")
plt.title("EMI Months Distribution")
plt.legend(emi_months_dist.index, loc="best")
plt.show()
```

```python
# 10.How does the interest rate vary across different brands
brand_interest_rate = carsdf.groupby("Brand")["Interest_Rate(%)"].mean().round(2)
print(brand_interest_rate)

plt.figure(figsize=(20,7))
plt.plot(brand_interest_rate)
plt.title("Interest Rate by Brand")
plt.xlabel("Brand")
plt.ylabel("Interest Rate")
plt.grid(True,axis="y",linestyle="--")
plt.show()
```

```python
# 11.Which state shows the highest EMI burden (average EMI_Amount)?
avg_emi_state = carsdf.groupby("State")["EMI_Amount"].mean().sort_values(ascending=False).round(2)
print(avg_emi_state)

plt.figure(figsize=(20,7))
sns.barplot(avg_emi_state,palette="Greens",errorbar=None)
plt.title("Average EMI Amount by State")
plt.xlabel("State")
plt.ylabel("Average EMI Amount")
plt.show()
```

```python
# 12.	Compare sales trends between urban (metros) vs. non-urban cities.
metro_cities = ["New Delhi", "Mumbai", "Kolkata", "Chennai", "Bengaluru", "Hyderabad", "Pune"]
carsdf["City_Type"] = np.where(carsdf["City"].isin(metro_cities), "Urban", "Non-Urban")

sales_trend = carsdf.groupby("City_Type")["Final_Price"].sum()
print(sales_trend)
```

```python
# 13.	Which state has the lowest average final price?
low_avg_price_state = carsdf.groupby("State")["Final_Price"].mean().sort_values(ascending=True).round(2).head()
print(low_avg_price_state)
```

```python
# 14.	Is there a difference in sales based on fuel type across state?
fuel_type_sales = carsdf.groupby(["State","Fuel_Type"])["Final_Price"].sum().reset_index()
print(fuel_type_sales)

plt.figure(figsize=(20,10))
sns.lineplot(data=fuel_type_sales,
             x="State",
             y="Final_Price",
             hue="Fuel_Type",
             palette="deep"
             )
plt.title("Sales by Fuel Type and State")
plt.xlabel("State")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

# **Car Features Analysis**

```python
# 15.	What is the most popular car color among buyers?
car_color = carsdf["Color"].value_counts()
print(car_color)

plt.figure(figsize=(25,7))
plt.pie(car_color,labels=car_color.values, colors=plt.cm.tab10.colors  )
plt.title("Car Color Distribution")
plt.legend(car_color.index,loc="best")
plt.show()
```

```python
# 16.	Which transmission type (Manual/AMT/DCT) is most preferred?
transmission_count = carsdf["Transmission"].value_counts()
print(transmission_count)
```

```python
# 17.	Compare the sales of Petrol vs Diesel vs Electric cars in each state.
sales_comparession = carsdf.groupby("Fuel_Type")["Final_Price"].sum().sort_values(ascending=False)
sales_comparession_filter = sales_comparession.loc[["Petrol", "Diesel", "Electric"]]
print(sales_comparession_filter)

plt.figure(figsize=(20,8))
sns.barplot(sales_comparession_filter,palette="deep")
plt.title("Sales Comparison by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Sales")
plt.show()
```

```python
# 18. Which model contributes the most to each brands sales?
model_contribution = carsdf.groupby(["Brand","Model"])["Final_Price"].sum().reset_index()

# Find the model with maximum sales per brand
top_model = model_contribution.loc[model_contribution.groupby("Brand")["Final_Price"].idxmax()]
top_model = top_model.sort_values(by="Final_Price", ascending=False)
print(top_model)


plt.figure(figsize=(20,8))
sns.barplot(x="Brand",y="Final_Price",hue="Model",data=top_model)
plt.title("Top Model by Brand")
plt.xlabel("Brand")
plt.ylabel("Sales")
plt.grid(axis="y",linestyle = "--", alpha = 0.3)
plt.show()
```

# **Profitability and Business Insights**

```python
# 19. Revenue by Brand
brand_revenue = carsdf.groupby("Brand")["Final_Price"].sum().sort_values(ascending=False)

brand_revenue.plot(
    kind="pie",
    title="Revenue By Brand",
    figsize=(20,6)
)
plt.show()
brand_revenue
#brand_revenue
```

```python
# 20.Which car model brought in the highest profit margin (Final_Price  Ex_Showroom_Price)?
model_profit = carsdf.groupby(["Brand","Model"])["Final_Price"].sum() - carsdf.groupby(["Brand","Model"])["Ex_Showroom_Price"].sum()

brand_model_profit = model_profit.loc[model_profit.groupby("Brand").idxmax()]
brand_model_profit = brand_model_profit.sort_values(ascending=False).reset_index()
brand_model_profit = brand_model_profit.rename(columns={0: "Profit"})
print(brand_model_profit)


plt.figure(figsize=(20,8))
sns.barplot(x="Brand",y="Profit",hue="Model",data=brand_model_profit)
plt.title("Top Model by Brand")
plt.xlabel("Brand")
plt.ylabel("Sales")
plt.grid(axis="y",linestyle = "--", alpha = 0.3)
plt.show()
```

```python
# 21.Do higher discounts correlate with higher sales volume?
discount_sales = carsdf.groupby("Discount").size().reset_index(name="Count")
print(discount_sales)

plt.figure(figsize=(20,8))
sns.scatterplot(x="Discount",y="Count",data=discount_sales)
plt.title("Discount vs. Sales Volume")
plt.xlabel("Discount")
plt.ylabel("Sales Volume")
plt.grid(axis="y",linestyle="--",alpha=0.5)
plt.show()
```

```python
# 22.Which brand is most loan-dependent (highest % of financed sales)?
loan_dependent_brand = carsdf.groupby("Brand")["Loan_Amount"].sum()/carsdf.groupby("Brand")["Final_Price"].sum() * 100
loan_dependent_brand = loan_dependent_brand.sort_values(ascending=False).head(5)
print(loan_dependent_brand)

plt.figure(figsize=(20,8))
plt.plot(loan_dependent_brand,marker="o")
plt.title("Loan-Dependent Brands")
plt.xlabel("Brand")
plt.ylabel("Loan %")
plt.grid(axis="y",linestyle="--",alpha=0.5)
plt.show()
```

# **Time Series & Seasonal Analysis**

```python
# 23.How do sales vary by month of 2024

# converting Sale_Date Column datatype to datetime datatype
carsdf["Sale_Date"]= pd.to_datetime(carsdf["Sale_Date"])

# Getting Monthly sales
monthly_sales = carsdf.groupby(carsdf["Sale_Date"].dt.month)["Final_Price"].sum()
monthly_sales.index = ["Jan","Feb","March","April","May","June","July","Aug","Sept","Oct","Nov","Dec"]
print(monthly_sales)

plt.figure(figsize=(20,8))
plt.plot(monthly_sales,marker="o",color ="Red")
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(axis="y",linestyle="--",alpha=0.4)
plt.show()
```

```python
# 24.How do sales vary by Quater of 2024
quater_sales = carsdf.groupby(carsdf["Sale_Date"].dt.quarter)["Final_Price"].sum()
qauter_sales_index = ["Q1","Q2","Q3","Q4"]
print(quater_sales)

plt.figure(figsize=(20,8))
sns.barplot(x=qauter_sales_index,y=quater_sales.values,color="Green")
plt.title("Quaterly Sales")
plt.xlabel("Quater")
plt.ylabel("Sales")
plt.grid(axis="y",linestyle="--",alpha=0.4)
plt.show()
```

```python
# 25.Are discounts more common in certain months
discounts_month = carsdf.groupby(carsdf["Sale_Date"].dt.month)["Discount"].sum()
discounts_month.index = ["Jan","Feb","March","April","May","June","July","Aug","Sept","Oct","Nov","Dec"]
print(discounts_month)

plt.figure(figsize=(20,8))
plt.plot(discounts_month,marker ="o")
plt.title("Discount by Month")
plt.xlabel("Month")
plt.ylabel("Discount")
plt.grid(axis="y",linestyle="--",alpha=0.4)
plt.show()
```
